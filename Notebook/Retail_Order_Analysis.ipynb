#### By observing the data we noticed that there are some value in 'Ship Mode' column that needs to be treated as null values 

```python
## we can use 'na_values' parameter to treat specific values as null values 
df = pd.read_csv("orders.csv",na_values = ['Not Available','unknown'])
df.head(10)                
```

```python
# checking 'Ship Mode' values again 
df['Ship Mode'].unique()
```

#### 'Not Available' & 'Unknown' are converted to null values

```python
df.columns
```

#### for good practice your coulumns should be in lower case and instead of spaces underscore should use 

```python
# replacing column names
df.columns = df.columns.str.lower() ## converting all columns to lower case 
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.head()
```

#### Data have 'Discount Percent' column but don't have column that tells exact dicounted price, so, er are creating 'Discount' column 

```python
df['Discount'] = df['list_price'] * df['discount_percent']/100
```

```python
df.head()
```

```python
df.rename(columns={'Discount':'discount'},inplace=True)
```

```python
df.columns
```

#### Also we need 'selling_Price' & 'profit' column

```python
df['selling_price'] = df['list_price'] - df['discount']
df['profit'] = df['selling_price'] - df['cost_price']
```

```python
df['selling_price']
```

```python
df.head()
```

#### Now we don't require 'list_price','discount_percent'&'discount' columns

```python
df.drop(columns = ['list_price','discount_percent', 'discount'],inplace = True)
```

```python
# order_time column have 'object datatype' need to convert it into 'datetime'

df['order_date'] = pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
```
